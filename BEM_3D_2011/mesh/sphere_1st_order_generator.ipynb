```python
import gmsh
import sys

# Initialize Gmsh
gmsh.initialize()
gmsh.model.add("sphere_surface")

# Define the sphere parameters
radius = 1.0
sphere_center = [0, 0, 0]

# Create a sphere surface
gmsh.model.occ.addSphere(sphere_center[0], sphere_center[1], sphere_center[2], radius)

# Synchronize the model with Gmsh's internal data structure
gmsh.model.occ.synchronize()

# Generate a 2D mesh (surface mesh)
gmsh.model.mesh.generate(2)

# Save the mesh to an STL file
gmsh.write("sphere_1st_order.stl")

# Finalize the Gmsh API
gmsh.finalize()
```

```python
import gmsh
import math
import os

def create_second_order_sphere(refinement_levels=4):
    # Initialize the Gmsh API
    gmsh.initialize()
    gmsh.model.add("SecondOrderSphere")

    # Define a sphere using the built-in sphere primitive
    sphere = gmsh.model.occ.addSphere(0, 0, 0, 1)  # Center (0, 0, 0) and radius 1
    gmsh.model.occ.synchronize()

    # Add a physical group for the surface of the sphere
    gmsh.model.addPhysicalGroup(2, [sphere], tag=1)
    gmsh.model.setPhysicalName(2, 1, "Sphere Surface")

    # Generate second-order mesh and refinement
    gmsh.option.setNumber("Mesh.ElementOrder", 2)  # Set second-order elements
    gmsh.option.setNumber("Mesh.RecombineAll", 1)  # Use recombination for quadrangles
    gmsh.option.setNumber("Mesh.Algorithm", 6)    # 3D meshing algorithm (Delaunay)

    # Create multiple refinement levels
    for i in range(refinement_levels):
        mesh_size = 0.5 / (2 ** i)  # Refinement factor
        gmsh.option.setNumber("Mesh.MeshSizeMax", mesh_size)
        gmsh.option.setNumber("Mesh.MeshSizeMin", mesh_size)

        # Generate the mesh
        gmsh.model.mesh.generate(3)

        # Save the mesh to a file
        filename = f"sphere_refinement_level_{i + 1}.msh"
        gmsh.write(filename)
        print(f"Mesh refinement level {i + 1} saved to {filename}")

    # Finalize the Gmsh API
    gmsh.finalize()

# Run the function to create and refine the sphere meshes
create_second_order_sphere()
```
